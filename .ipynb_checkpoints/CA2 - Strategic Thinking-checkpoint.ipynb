# Creating Features. 

```python
games['month'] = games['Release_Date'].dt.month
```

```python
games['year'] = games['Release_Date'].dt.year
```

```python
games.head()
```

```python
# Created a function to replace K to the numeric value times 1000. 

def replace_k(games):
    if 'K' in games:
        return float(games.replace('K', '')) * 1000
    else: 
        return float(games)
    
```

```python
columns_game = ['Wishlist', 'Backlogs', 'Playing', 'Plays', 'Number_of_Reviews', 'Times_Listed']
```

```python
for column in columns_game:
    games[column] = games[column].apply(replace_k)
```

# Applying Sentiment my review

    As Always we need to go throught the process of cleaning data, however, we dealing with the most complicate situation in Data Analytics Text! To be able to continue my process I will be using some methods regarding to extracting the sentimental in the feature Reviews. 

1. Stop words
2. Sentiment

```python
from nltk.corpus import stopwords
```

```python
stop = stopwords.words("english")
```

```python
games["stopwords"]= games["Reviews"].apply(lambda x:len([x for x in x.split() if x in stop]))
```

```python
games[["Reviews", "stopwords"]].head()
```

```python
games["hashtags"] = games["Reviews"].apply(lambda x: len([word for word in x.split() if "#" in word]))
```

```python
games[["Reviews", "stopwords", "hashtags"]].head()
```

```python
games["numerics"]= games["Reviews"].apply(lambda x:len([x for x in x.split() if x.isdigit()]))
```

```python
games[["Reviews", "numerics"]].head()
```

```python
from textblob import TextBlob
```

```python
games['Reviews'].apply(lambda x: TextBlob(x).sentiment)
```

```python
games['sentiment'] = games['Reviews'].apply(lambda x: TextBlob(x).sentiment[0])
```

```python
column_3 = ['numerics','hashtags', 'stopwords', 'Reviews' ]
```

# Experiments

    I've replaced the values of my target variable to have binary variable. My logic was if the game rating is greater than 3.5 I assign to 0 if is less than 3.5 I assign to 1. 
    My hypothesis is that if I change my target variable to binary variable with 0 and 1 in instead of random ratings I will be able to get a good score in my machine learning algorithm, however, to be sure we need to test. 
    I've already test which machine learning algorithm is the best fit to my Dataset, follow the results. 

##### We succeed to improve our results changing to 3.5 to 4.0, because, when I created our code I found out that I had desbalanced data due to I have too many greater than 3.5 I will desmostrade the results below. 

##### We improved our results in Random Forest Classifier, Gradient Boosting Classifier, however, in Logistic regression and SVC there was a drop in the results as you can see above. 
###### In order to get the highest results I will apply my experiment here. 

```python
# My ideia is try to change the values to be able to predict each game is more likely to get a rating of 3.5 is really good rating.  
def rating_values(data):
    new_rating = []
    for rating in data["Rating"]:
        if rating >= 4.0:
            new_rating.append(0)
        elif rating < 4.0:
            new_rating.append(1)
    return new_rating
```

```python
new_rating = rating_values(data=games)
games["Rating"] = new_rating
```
